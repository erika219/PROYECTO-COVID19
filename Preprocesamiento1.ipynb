```
%%time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#datos etiquetados
#datos no etiquetados
#Tecnicas clasificacion
#T. Asociación
#T. Agrupacion, clusters
#Aprendizaje supervisado y no supervisado

from sklearn.cluster import KMeans
def crear(x):
  df1 = pd.DataFrame(np.random.randint(0,x,size=(x, 2)), columns=list('AB'))
  modelo=KMeans(n_clusters=5).fit(df1)
  modelo.fit(df1)
  centros=modelo.cluster_centers_
  x=modelo.inertia_
  plt.scatter(df1['A'],df1['B'],c=modelo.labels_.astype(float), s=50,alpha=0.5)
  plt.scatter(centros[:,0],centros[:,1],c='r',s=150)
  plt.show()
crear(1000)
```
